```python
import tkinter as tk
from tkinter import filedialog
import numpy as np
import os
from brukeropusreader import read_file

def calculate_snr(signal, noise):
    """
    Berechnet das Signal-to-Noise Ratio (SNR).
    :param signal: Numpy Array des Signals.
    :param noise: Numpy Array des Rauschens.
    :return: SNR-Wert.
    """
    signal_power = np.mean(signal ** 2)
    noise_power = np.mean(noise ** 2)
    snr = 10 * np.log10(signal_power / noise_power)
    return snr

def process_files():
    root = tk.Tk()
    root.withdraw()  # Verstecke das Hauptfenster
    # Öffne den Dateidialog und hole die Dateipfade
    file_paths = filedialog.askopenfilenames()
    for file_path in file_paths:
        if file_path:
            # Lese die OPUS-Datei
            opus_data = read_file(file_path)
            file_name = os.path.basename(file_path)
            absorbance = opus_data['AB']

            start_wn = opus_data['AB Data Parameter']['FXV']
            end_wn = opus_data['AB Data Parameter']['LXV']
            num_points = opus_data['AB Data Parameter']['NPT']
            num_points = len(absorbance)
            wavenumbers = np.linspace(start_wn, end_wn, num_points)
            assert len(wavenumbers) == len(absorbance), "Wavenumbers and absorbance arrays must be the same length."

            # Hier fügen wir eine einfache Rauschberechnung hinzu, indem wir einen Teil des Spektrums als Rauschen betrachten
            # Dies ist nur ein Beispiel und sollte an Ihre spezifischen Daten angepasst werden
            noise_region = absorbance[:100]  # Nehmen wir die letzten 50 Punkte als Rauschen
            snr_value = calculate_snr(absorbance, noise_region)
            print(f"SNR for {file_name}: {snr_value} dB")

process_files()
```
