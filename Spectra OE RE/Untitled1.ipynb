```python
import numpy as np
import os

# Beispiel: Pfad zum Ordner mit den normalisierten Spektren
folder_path = "/Users/leokavallaris/FTIR ATR/Spectra OE RE/normalized_OE_RE"


# Durch die CSV-Dateien im Ordner iterieren
for file_name in os.listdir(folder_path):
    if file_name.endswith("_normalized.csv"):
        # Lese die normalisierten Spektraldaten ein
        file_path = os.path.join(folder_path, file_name)
        normalized_absorbance = np.loadtxt(file_path, delimiter=',')
        
        # Erzeuge die Wellenzahlen
        start_wn = 400 # Anpassen, falls bekannt
        end_wn = 5000  # Anpassen, falls bekannt
        num_points = len(normalized_absorbance)
        wavenumbers = np.linspace(start_wn, end_wn, num_points)
        
        # Bereich der Wellenlängen definieren (1800-900)
        wl_start = 1800
        wl_end = 900
        
        # Indizes für den Wellenlängenbereich erhalten
        idx_start = np.where(wavenumbers >= wl_start)[0][0]
        idx_end = np.where(wavenumbers <= wl_end)[0][-1]
        
        # Ausschnitt der normalisierten Absorption für den Bereich auswählen
        selected_wavenumbers = wavenumbers[idx_start:idx_end+1]
        selected_absorbance = normalized_absorbance[idx_start:idx_end+1]
        
        # Pfad zum Ordner, in dem die ausgewählten Daten gespeichert werden sollen
        save_folder = "path_to_normalized_OE_RE_1800_9002"
        
        # Überprüfen, ob der Speicherordner existiert, wenn nicht, erstellen
        if not os.path.exists(save_folder):
            os.makedirs(save_folder)
        
        # Dateipfad für die ausgewählten Daten
        save_file_path = os.path.join(save_folder, f"{file_name[:-16]}_1800_900_normalized.csv")
        
        # Speichern der ausgewählten Daten in einer CSV-Datei
        np.savetxt(save_file_path, np.column_stack((selected_wavenumbers, selected_absorbance)), delimiter=',', header='Wavenumber (cm⁻¹),Normalized Absorbance', comments='')
        
        print(f"Datei '{file_name[:-16]}_1800_900_normalized.csv' wurde erfolgreich gespeichert.")
```
